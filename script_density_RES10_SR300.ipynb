# Bay Features

## GeoNames

```python
geonames_bays = DiscoBall(res=res)
print(geonames_bays)
```

```python
geonames_bays.load_boundaries('data\ocean_borders.geojson')
```

```python
if os.path.exists(f'computed_layers\geonames_bays_RES{res}_SR{sr}.txt'):    
    geonames_bays.load_layer('density', f'computed_layers\geonames_bays_RES{res}_SR{sr}.txt')
    
else:
    geonames_bays.add_multiple_points_to_raster('data\point_datasets\GeoNames_classed_points.json', key='Bay')    
    geonames_bays.compute_point_density_layer(sr)
    geonames_bays.save_layer('density', f'computed_layers\geonames_bays_RES{res}_SR{sr}.txt')
```

```python
geonames_bays_bins = geonames_bays.classify_layer('density', 5)
```

```python
geonames_bays.create_image_from_layer('density_classified', f'maps\geonames_bays_RES{res}_SR{sr}.png')
```

```python
display(Image(filename=f'maps\geonames_bays_RES{res}_SR{sr}.png'))
```

## MarineRegions

```python
marineregions_bays = DiscoBall(res=res)
print(marineregions_bays)
```

```python
marineregions_bays.load_boundaries('data\ocean_borders.geojson')
```

```python
if os.path.exists(f'computed_layers\marineregions_bays_RES{res}_SR{sr}.txt'):    
    marineregions_bays.load_layer('density', f'computed_layers\marineregions_bays_RES{res}_SR{sr}.txt')
    
else:
    marineregions_bays.add_multiple_points_to_raster('data\point_datasets\MarineRegions_classed_points.json', key='Bay')    
    marineregions_bays.compute_point_density_layer(sr)
    marineregions_bays.save_layer('density', f'computed_layers\marineregions_bays_RES{res}_SR{sr}.txt')
```

```python
legend_bins = marineregions_bays.classify_layer('density', 5, bins=geonames_bays_bins)
```

```python
marineregions_bays.create_image_from_layer('density_classified', f'maps\marineregions_bays_RES{res}_SR{sr}.png')
```

```python
display(Image(filename=f'maps\marineregions_bays_RES{res}_SR{sr}.png'))
```

#### Color Legend

```python
print(f'Color legend bins: {legend_bins}')
```
